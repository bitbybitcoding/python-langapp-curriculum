# Let's learn how to analyse a text file!

## An important function for dealing with files in Python is `open()`.

### There are 4 primary ways to interact with files:
- "w", which is for writing/overwriting
- "r", which is for reading
- "a", which is for appending content
- "x", which is for creating files

### Moreover, the 2 types of files you may deal with are
- "t", which are text files
- "b", which are binary files such as images

# Section 3 : Advanced Reading
Below shows 3 alternative ways to read, allowing you to specify what part of the file you want read, and how the data returned should be formatted. This can help a lot in data cleaning.

<table>
  <thead>
    <tr>
      <th>Method</th>

    </tr>

    </tr>
  </tbody>
</table>

*Note: `readline()` and `readlines()` will both return any trailing `"\n"` at the end of each line.*

# Section 4 : Advanced Writing
Below shows the 2 main ways to write to a file.

<table>
  <thead>
    <tr>
      <th>Method</th>
      <th>Action</th>
      <th>Expected Input Type</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>f.write()</code></td>
      <td>Writes a specific string to the file.</td>

    </tr>
    <tr>

  </tbody>
</table>

*Note: For `writelines()`, Python won't help you add `"\n"` to the end of the each string in the list, you have to add it yourself.*
